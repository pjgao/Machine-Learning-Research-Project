```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
X_train=pd.read_csv('/Users/zhilinyang/Desktop/down_train_x.csv')
len(X_train)
```

```python
X_train=X_train.iloc[:,1:]
X_train.head()
```

```python
y=[0 for i in range(0,1000)]+[1 for i in range(0,458)]
y_train=np.array(y)
y_train
```

```python
y_train=pd.DataFrame(y_train)
y_train
```

```python
dataset=pd.concat([X_train,y_train],axis=1,ignore_index=True)
dataset.head()
```

```python
dataset=dataset.sample(frac=1)
dataset.head()
```

```python
X_train=dataset.iloc[:,0:24]
y_train=dataset.iloc[:,25]
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
scores=cross_val_score(estimator=lr,scoring='precision',X=X_train,y=y_train,cv=10,n_jobs=2)
print('lr replaced precision: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
scores=cross_val_score(estimator=lr,scoring='recall',X=X_train,y=y_train,cv=10,n_jobs=2)
print('lr replaced recall: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
scores=cross_val_score(estimator=lr,scoring='f1',X=X_train,y=y_train,cv=10,n_jobs=2)
print('lr replaced f1: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
```

```python
from sklearn.metrics import make_scorer,precision_score,recall_score,f1_score
f1_scorer=make_scorer(f1_score,pos_label=0)
precision_scorer=make_scorer(precision_score,pos_label=0)
recall_scorer=make_scorer(recall_score,pos_label=0)
```

```python
scores=cross_val_score(estimator=lr,scoring=precision_scorer,X=X_train,y=y_train,cv=10,n_jobs=2)
print('lr healthy precision: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
scores=cross_val_score(estimator=lr,scoring=recall_scorer,X=X_train,y=y_train,cv=10,n_jobs=2)
print('lr healthy recall: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
scores=cross_val_score(estimator=lr,scoring=f1_scorer,X=X_train,y=y_train,cv=10,n_jobs=2)
print('lr healthy f1: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_range=[0.0001,0.001,0.01,0.1,1.0,10.0,100.0,1000.0]
param_grid=[{'C':param_range}]
gs=GridSearchCV(estimator=lr,param_grid=param_grid,cv=10,n_jobs=-1)
gs=gs.fit(X_train, y_train)
print(gs.best_score_)
```

```python
from rgf.sklearn import RGFClassifier
RGF = RGFClassifier(max_leaf=10000,l2=1.0,sl2=0.005,test_interval=100,verbose=True)
```

```python
from sklearn.learning_curve import learning_curve
import matplotlib.pyplot as plt
train_sizes, train_scores, test_scores =\
learning_curve(estimator=RGF, X=X_train, y=y_train, 
               train_sizes=np.linspace(0.1,1.0,10),cv=10,n_jobs=1)
train_mean=np.mean(train_scores, axis=1)
test_mean=np.mean(test_scores, axis=1)
```

```python
plt.plot(train_sizes, train_mean, color='blue', marker='o',label='training accuracy')
plt.plot(train_sizes, test_mean, color='red', marker='s',label='validation accuracy')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()
```
